```python
import pandas as pd
import copy
import math
import random
import sklearn
```

```python
# import dataframe
data_players = pd.read_csv('players.csv')
data_players = data_players.fillna(0)
data_players.shape
```

```python
data_players.head()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5, random_state=0).fit(data_players.drop(columns=['player_name']))
kmeans.labels_
```

```python
data_players['new_cat'] = kmeans.labels_
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
data_players.sort_values(by='new_cat')
```

```python
data_merge = data_players[['player_name','new_cat']]
data_merge
```

```python
data = pd.read_csv('players_old.csv')
```

```python
pd.merge(data, data_merge, on='player_name').to_csv('players_kmeans.csv',index=False)
```
